# **Problem**
## Tidness issue
 ### The issue with the data is that locations (Yangon, Naypyitaw, Mandalay) are represented as columns, which creates redundant columns for each location.

# **Code**

```python
df_melted = pd.melt(df,
                    id_vars=['Invoice ID', 'Branch', 'Customer type', 'Gender', 'Product line',
                             'Unit price', 'Quantity', 'Tax 5%', 'Total', 'Date', 'Time',
                             'Payment', 'Rating'],
                    value_vars=['Yangon', 'Naypyitaw', 'Mandalay'],
                    var_name='City', value_name='Presence')

# Filter to keep only rows where 'Presence' is 1
df = df_melted[df_melted['Presence'] == 1].drop(columns=['Presence'])

# Rearrange the columns to make 'City' the third column
df = df[['Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line',
                     'Unit price', 'Quantity', 'Tax 5%', 'Total', 'Date', 'Time', 'Payment',
                     'Rating']]
```

# **Problem**
## Unit price column not in correct data tybe

## **Why ?**
### becuse it is object ( have numeric and non-numeric )

##**Check non numeric values**

```python
# Step 1: Identify the rows that initially contained non-numeric values (e.g., "USD")
non_numeric = df[~df['Unit price'].str.match(r'^\d+\.?\d*$', na=False)]

# Print the original rows containing "USD" or other non-numeric characters
print("Original rows with non-numeric values:")
non_numeric[['Unit price']]
```

### Step 1: Remove "USD" or other non-numeric characters from the 'Unit price'


```python
df.loc[non_numeric.index, 'Unit price'] = df.loc[non_numeric.index, 'Unit price'].replace({'[^\d.]': ''}, regex=True)
```

### Step 2: Convert the 'Unit price' column to float


```python
# Step 1: Convert the 'Unit price' column to float
df['Unit price'] = pd.to_numeric(df['Unit price'], errors='coerce')
```

#**Test**

```python

# Step 3: Check the cleaned values for these specific rows
cleaned_values = df.loc[non_numeric.index, 'Unit price']

# Print the cleaned values to verify that "USD" was removed
print("\nCleaned values for the same rows after removing 'USD':")
cleaned_values
```
